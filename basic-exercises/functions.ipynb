### Functions

Define a function 'dictionary_creator' to take in any two lists and return a dictionary between list1 and list2. If the two lists are not the same size, return the word "Error" instead.


```python
def dictionary_creator(list1, list2):
    
    ### dictionary_creator(['a', 'b', 'c'], [1,2,3]) should return {'a': 1, 'b': 2, 'c': 3}
    ### dictionary_creator(['monkeys', 'pirates', 'captains'], [1,2]) should return "Error"
    
    new_dict = {}
    # your code here
    if len(list1) != len(list2):
        return "Error"
    else:
        for index, i in enumerate(list1):
            new_dict[i] = list2[index]
            
    return new_dict
    
```
